```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
import warnings

from IPython.display import display
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# Importing dataset
mice = pd.read_csv("Data_Cortex_Nuclear.csv")
mice = mice.drop_duplicates()
display(mice.sample(5))
```

```python
pp.ProfileReport(mice)
```

```python
print((mice[['ELK_N', 'Bcatenin_N','BAD_N', 'BCL2_N', 'pCFOS_N', 'H3AcK18_N', 'EGR1_N', 'H3MeK4_N']].isnull().sum()))
```

The missing values in ELK_N, Bcatenin_N nd pCFOS_N are very little compared to the size of the of the data set. But the missing values for other variables are quite significant where they are more than 20% of the data set. So it's better to imputate the missing values instead of removing them and I choose to imputate them 

```python
# fill missing values with mean column values
mice.fillna(mice.mean(), inplace=True)
print((mice[['ELK_N', 'Bcatenin_N','BAD_N', 'BCL2_N', 'pCFOS_N', 'H3AcK18_N', 'EGR1_N', 'H3MeK4_N']].isnull().sum()))
```
